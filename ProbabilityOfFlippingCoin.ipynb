```python
import random
import matplotlib.pyplot as plt
```

```python
tr = int(input("Enter number of times to flip the coin: "))
coin = {'H':0,'T':0}
prob_H_list = []
prob_T_list = []
for i in range(1,tr+1):
  rdint = random.randint(0,1)
  if(rdint == 0):
    coin['H'] = coin['H'] + 1
  else:
    coin['T'] = coin['T'] + 1
  prob_H_list.append(coin['H'] / i)
  prob_T_list.append(coin['T'] / i)


plt.figure(figsize=(10, 5))
plt.plot(range(1, tr + 1), prob_H_list, label='Probability of Heads (H)', color='blue')
plt.plot(range(1, tr + 1), prob_T_list, label='Probability of Tails (T)', color='orange')
plt.axhline(y=0.5, color='gray', linestyle='--', label='Expected Probability = 0.5')


plt.title("Probability of Heads and Tails Over Trials")
plt.xlabel("Trial Number")
plt.ylabel("Probability")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
tr = int(input("Enter number of trials: "))
num_coins = int(input("Enter number of coins per trial: "))

coin = {'H': 0, 'T': 0}
prob_H_list = []
prob_T_list = []

total_flips = 0

for i in range(1, tr + 1):
    for _ in range(num_coins):
        rdint = random.randint(0, 1)
        if rdint == 0:
            coin['H'] += 1
        else:
            coin['T'] += 1
        total_flips += 1

    # Update probabilities after this trial (cumulative)
    prob_H_list.append(coin['H'] / total_flips)
    prob_T_list.append(coin['T'] / total_flips)

# Visualization
plt.figure(figsize=(10, 5))
plt.plot(range(1, tr + 1), prob_H_list, label='Probability of Heads (H)', color='blue')
plt.plot(range(1, tr + 1), prob_T_list, label='Probability of Tails (T)', color='orange')
plt.axhline(y=0.5, color='gray', linestyle='--', label='Expected Probability = 0.5')

plt.title(f"Probability of Heads and Tails Over {tr} Trials ({num_coins} coins per trial)")
plt.xlabel("Trial Number")
plt.ylabel("Cumulative Probability")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
