```python
import math
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
n = int(input("Enter sample size: "))
Sample_Mean = float(input("Enter Sample Mean: "))
Sample_SD = float(input("Enter Sample SD: "))
CI = int(input("Enter Confidence Interval: "))/100
z_score = stats.norm.ppf(1 - (1 - CI) / 2)

SE = z_score*(Sample_SD/math.sqrt(n))
print("Standard Error is: ", SE)

Lower_CI = Sample_Mean - SE
Upper_CI = Sample_Mean + SE

print(f"The {CI}% confidence interval is: [{Lower_CI:.3f}, {Upper_CI:.3f}]")

x = np.linspace(Sample_Mean - 4 * Sample_SD, Sample_Mean + 4 * Sample_SD, 1000)
y = stats.norm.pdf(x, Sample_Mean, Sample_SD / math.sqrt(n))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Normal Distribution')
plt.fill_between(x, y, where=((x >= Lower_CI) & (x <= Upper_CI)), color='green', alpha=0.3, label=f'{CI * 100}% Confidence Interval')
plt.axvline(x=Sample_Mean, color='blue', linestyle='--', label='Sample Mean')
plt.axvline(Sample_Mean, color='red', linestyle='--', label='Sample Mean')
plt.axvline(Lower_CI, color='orange', linestyle='--', label=f'Lower CI ({Lower_CI:.3f})')
plt.axvline(Upper_CI, color='orange', linestyle='--', label=f'Upper CI ({Upper_CI:.3f})')
plt.title(f"Confidence Interval for %{CI*100:.0f} Confidence")
plt.xlabel("Values")
plt.ylabel("Density")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Show the plot
plt.show()
```
