```python
import pandas as pd
```

```python
DF = pd.read_csv('daily-minimum-temperatures-in-me.csv')
```

```python
"""
2.Convert Date Column
  Convert the date column to a datetime object
  Verify the data type of the column
"""
```

```python
import datetime as dt
from datetime import datetime,date,time,timedelta
```

```python
DF['Date'] = pd.to_datetime(DF['Date'])
type(DF['Date'][0])
```

```python
"""
3.Extract Features
  -Extract the year, month, and day from the date column and
   add them as new columns.

  -Create a column for the day of the week (e.g., Monday, Tuesday).
"""
```

```python
DF['year'] = DF['Date'].dt.year
DF['month'] = DF['Date'].dt.month
DF['day'] = DF['Date'].dt.day
DF[['year','month','day']].head()
```

```python
DF['dayOfWeek'] = DF['Date'].dt.day_name()
DF['dayOfWeek'].head()
```

```python
"""
4.Filter Data
  - Filter the dataset to include only the temperatures
    from a specific year (e.g., 1985)

  - Count how many days in 1985 had a temperature below
   a certain threshold (e.g., 10°C)
"""
```

```python
DF[DF['year'] == 1985]
```

```python
DF['Temperature'] = pd.to_numeric(DF['Daily minimum temperatures'], errors='coerce')
```

```python
DF[(DF['year'] == 1985) & (DF['Temperature'] < 10)]['day'].count()
```

```python
"""
5.Analyze Time Periods
 - Group the data by year and calculate the average minimum
   temperature for each year.

 - Find the year with the highest and lowest average minimum temperatures.
"""
```

```python
DF.groupby('year')['Temperature'].min()
```

```python
DF.groupby('year')['Temperature'].mean().idxmax()
```

```python
DF.groupby('year')['Temperature'].mean().idxmin()
```
