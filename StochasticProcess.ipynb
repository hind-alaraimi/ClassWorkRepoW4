```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
initial_population = 100
birth_rate = 0.05  # 10% chance of birth per time step
death_rate = 0.05 # 5% chance of death per time step
time_steps = 200  # Number of time steps

# Initializing the population array
population = [initial_population]

for _ in range(time_steps):
    current_population = population[-1]

    # Calculating number of births and deaths
    births = np.random.poisson(current_population * birth_rate)
    deaths = np.random.poisson(current_population * death_rate)

    # Updating the population
    new_population = max(0, current_population + births - deaths)
    population.append(new_population)

# Plotting the population growth
plt.figure(figsize=(12, 6))
plt.plot(population, color="seagreen")
plt.title("Stochastic Process: Population Growth (Birth and Death Process)")
plt.xlabel("Time Steps")
plt.ylabel("Population")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()

print(population)
```
